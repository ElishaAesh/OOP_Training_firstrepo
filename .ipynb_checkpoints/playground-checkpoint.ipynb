```python
counter = 0; total = 0
grade = float(input("Enter grade(or negative value to the end): "))
while grade >= 0:
    total += grade
    counter += 1 
    grade = float(input("Enter grade(or negative value to the end): "))
print("The average grade is", total / counter)
```

```python
num_grades = int(input("Enter the number of grades"))
if num_grades > 0:
    first_grade = float(input("Enter grade # 1:"))

    total = first_grade
    max_grade = first_grade
    count = 1

    while count < num_grades:
        grade = float(input(f"Enter grade # {count + 1} : "))
        total = total + grade

        if grade > max_grade:
            max_grade = grade
        count = count + 1

    average = total/num_grades
    print("\n Results: ")
    print("Average grade: ", average)
    print("Maximum grade:", max_grade)
else:
    print("No grade were Entered")
```

```python
is_leap_year = False
input_year = int(input("Enter year: "))
if input_year % 4 == 0:
    if input_year % 100 == 0:
        if input_year % 400 == 0:
            is_leap_year = True 
    else:
        is_leap_year = True
if is_leap_year:
    print(input_year, "is a leap year")
else:
    print(input_year, "is not a leap year")
```

```python
# ----------------------------------------
# Python Cafe Ordering System
# ----------------------------------------

# Constants
COFFEE_PRICE = 3.00
TEA_PRICE = 2.50
HOT_CHOC_PRICE = 3.75

SUGAR_PRICE = 0.10
CREAM_PRICE = 0.50
SYRUP_PRICE = 0.75

DISCOUNT_THRESHOLD = 5.00
DISCOUNT_RATE = 0.10
TAX_RATE = 0.05


# Running totals
grand_total = 0
total_drinks = 0

another = "Y"

print("========================================")
print("          WELCOME TO PYTHON CAFE")
print("========================================")

while another == "Y":

    # ------------------------------
    # Beverage Menu
    # ------------------------------
    print("\n------------- BEVERAGE MENU -------------")
    print(f"{'1. Coffee':<25} ${COFFEE_PRICE:.2f}")
    print(f"{'2. Tea':<25} ${TEA_PRICE:.2f}")
    print(f"{'3. Hot Chocolate':<25} ${HOT_CHOC_PRICE:.2f}")
    print("-----------------------------------------")

    drink = int(input("Select beverage (1-3): "))

    if drink == 1:
        beverage_name = "Coffee"
        beverage_price = COFFEE_PRICE
    elif drink == 2:
        beverage_name = "Tea"
        beverage_price = TEA_PRICE
    else:
        beverage_name = "Hot Chocolate"
        beverage_price = HOT_CHOC_PRICE

    print("\nYou selected:", beverage_name)

    # ------------------------------
    # Extras Menu
    # ------------------------------
    print("\n--------------- EXTRAS MENU -------------")
    print(f"{'1. Sugar':<25} ${SUGAR_PRICE:.2f}")
    print(f"{'2. Cream':<25} ${CREAM_PRICE:.2f}")
    print(f"{'3. Syrup':<25} ${SYRUP_PRICE:.2f}")
    print("0. Finish Extras")
    print("-----------------------------------------")

    extras_total = 0
    extras_list = ""

    extra = int(input("Select extra (0-3): "))

    while extra != 0:

        if extra == 1:
            extras_total += SUGAR_PRICE
            extras_list += "Sugar "
            print("✓ Sugar added")

        elif extra == 2:
            extras_total += CREAM_PRICE
            extras_list += "Cream "
            print("✓ Cream added")

        elif extra == 3:
            extras_total += SYRUP_PRICE
            extras_list += "Syrup "
            print("✓ Syrup added")

        extra = int(input("Select extra (0-3): "))

    # ------------------------------
    # Calculations
    # ------------------------------
    subtotal = beverage_price + extras_total

    discount = 0
    if subtotal > DISCOUNT_THRESHOLD:
        discount = subtotal * DISCOUNT_RATE

    tax = (subtotal - discount) * TAX_RATE
    total = subtotal - discount + tax

    # ------------------------------
    # Order Summary
    # ------------------------------
    print("\n-----------------------------------------")
    print("              ORDER SUMMARY")
    print("-----------------------------------------")

    print(f"{'Beverage:':<20} {beverage_name}")
    print(f"{'Extras:':<20} {extras_list}")

    print()
    print(f"{'Beverage Price:':<20} ${beverage_price:.2f}")
    print(f"{'Extras Total:':<20} ${extras_total:.2f}")

    print("-----------------------------------------")
    print(f"{'Subtotal:':<20} ${subtotal:.2f}")
    print(f"{'Discount:':<20} ${discount:.2f}")
    print(f"{'Tax (5%):':<20} ${tax:.2f}")
    print("-----------------------------------------")
    print(f"{'Order Total:':<20} ${total:.2f}")
    print("-----------------------------------------")

    grand_total += total
    total_drinks += 1

    another = input("\nWould you like to order another drink? (Y/N): ")

# ------------------------------
# Final Receipt
# ------------------------------
print("\n========================================")
print("              FINAL RECEIPT")
print("========================================")

print(f"{'Total Drinks Ordered:':<30} {total_drinks}")
print("----------------------------------------")
print(f"{'GRAND TOTAL:':<30} ${grand_total:.2f}")
print("----------------------------------------")

print("Thank you for visiting Python Cafe!")
```

```python
SAFEST_CITIES = (
    "Tokyo, Japan",
    "Singapore, Singapore",
    "Osaka, Japan",
    "Amsterdam, Netherlands",
    "Sydney, Australia",
    "Toronto, Canada",
    "Washington D.C., United States",
    "Copenhagen, Denmark",
    "Seoul, South Korea",
    "Melbourne, Australia"
)

country = input("Enter the country to search for: ")

count = 0


for i in range(len(SAFEST_CITIES)):
    city = SAFEST_CITIES[i]
    
    if city.lower().endswith(country.lower()):
        print(f"Number {i+1} of 10: {city}")
        count += 1

print(f"There are {count} cities in {country} among the top 10 SAFEST cities in the world")
```

```python
students = []
scores = []

count = 1

while True:
    name = input(f"{count}. Student name: ")

    if name.upper() == "END":
        break

    score = float(input("   Score: "))

    students.append(name)
    scores.append(score)

    count += 1
    print()

total = sum(scores)
average = total / len(scores)

highest_score = max(scores)
highest_index = scores.index(highest_score)
highest_student = students[highest_index]

print(f"\nClass average score is {average:.1f}")
print(f"Highest score of {highest_score:.0f} achieved by {highest_student}!\n")

print("Student Name   Grade")
print("------------   -----")

for i in range(len(students)):
    print(f"{students[i]:<13} {scores[i]:.1f}")
```
